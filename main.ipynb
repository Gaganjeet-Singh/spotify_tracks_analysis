```python
import matplotlib.pyplot as plt

# Boxplot for song duration
plt.figure(figsize=(6,4))
plt.boxplot(df_copy['duration_sec'], vert=False, patch_artist=True,
            boxprops=dict(facecolor="lightgreen", color="green"),
            whiskerprops=dict(color="green"),
            capprops=dict(color="green"),
            medianprops=dict(color="red"))

plt.title("Boxplot of Song Duration (in Seconds)", fontsize=14)
plt.xlabel("Duration (seconds)", fontsize=12)
plt.show()
```

```python
Q1 = df_copy['duration_sec'].quantile(0.25)
Q3 = df_copy['duration_sec'].quantile(0.75)

IQR = Q3-Q1

lower_bound = Q1-(1.5*IQR)
upper_bound = Q3+(1.5*IQR)

print(f"LOWER BOUND : {lower_bound}, UPPER BOUND : {upper_bound}")
```

```python
df_filtered = df_copy[(df_copy['duration_sec'] >= lower_bound) &
                      (df_copy['duration_sec'] <= upper_bound)]

df_filtered.shape
```

```python
plt.hist(df_filtered['duration_sec'],bins=50,color ='green',edgecolor='white')
plt.title("Distribution Of Song Duration")
plt.xlabel("Duration (Seconds)")
plt.ylabel("Number Of Song")
plt.show()
```
